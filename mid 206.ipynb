```python
labels = ['Kanchanaburi', 'Rayong', 'Narathiwat']
sizes = [50, 100, 150]
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, shadow=True, autopct='%1.2f%%')
ax.axis('equal')  
ax.set_title('the provinces of Kanchanaburi, Rayong, and Narathiwat')
```

```python
explode = (50, 100, 150)

fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.2f%%',
       shadow=True,
       startangle=180)
ax.axis('equal')  
ax.set_title('the provinces of Kanchanaburi, Rayong, and Narathiwat')


plt.show()
```

```python
explode = (0.5, 0.1, 0.4)

fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.2f%%',
       shadow=True,
       startangle=180)
ax.axis('equal')  
ax.set_title('the provinces of Kanchanaburi, Rayong, and Narathiwat')


plt.show()
```
