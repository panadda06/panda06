```python
n = 10
pascal = np.zeros((n, n), dtype=int)
pascal[0, 0] = pascal[1, 0] = pascal[1, 1] = 1
for row in range(2, n):
    for col in range(0, row+1):
        pascal[row, col] = pascal[row-1, col-1] + pascal[row-1, col]
        print(pascal)
```
